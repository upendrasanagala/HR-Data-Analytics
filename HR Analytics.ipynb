```python
# create Age  New Column
bins = [20, 30, 40, 50, 60]
labels = ["21-30", "31-40", "41-50", "51-60"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
# salary Band

bins = [0, 500000, 800000, 1200000, 2000000]
labels = ["Low", "Medium", "High", "Very High"]

df["SalaryBand"] = pd.cut(df["Salary"], bins=bins, labels=labels)
```

```python
# experience level

def experience_level(exp):
    if exp <= 2:
        return "Fresher"
    elif exp <= 5:
        return "Junior"
    elif exp <= 10:
        return "Mid-Level"
    else:
        return "Senior"

df["ExperienceLevel"] = df["Experience"].apply(experience_level)
```

```python
# convert joing date to date time 
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"], errors="coerce")
```

```python
# tenure 

today = pd.Timestamp.today()

df["Tenure"] = ((today - df["JoiningDate"]).dt.days / 365).round(1)
```

```python
# save the cleaned dataset

df.to_csv("HR_Analytics_Cleaned.csv", index=False)
```
